ORIENTAÇÕES A SEREM RESPONDIDAS:

1. pegue somente os pokemons de grama lendários
2. pegue o pokemon com mais vida 
3. pegue os pokemons com mais de 80 de spAtk e menos de 120 spAtk
4. pegue a geração pokemon com maior numero de pokemons
5. pegue os pokemons com apenas um único tipo, quantos pokemons são e qual a media entre os tipos
6. pegue os pokemons que começam com j 
7. pegue o tipo mais comum de pokemon

```python
#importação dos modulos que vamos usar 
import pandas as pd
```

```python
#importação da tabela.csv
tabela_original = pd.read_csv('Pokemon.csv')
display(tabela_original)
```

```python
# 1 . pegue somente os pokemons de grama lendários

#copiando a tabela original
tabela = tabela_original.copy()

#lendo linha por linha dos tipos 1 e 2 e dos lendários
for linha in tabela.index:
    tipo1 = tabela.loc[linha, 'Type 1']
    tipo2 = tabela.loc[linha, 'Type 2']
    lendario = tabela.loc[linha, 'Legendary']
    
    #removendo os valores que não vamos usar, ou seja, os diferentes de grama
    if tipo1 != 'Grass' and tipo2 != 'Grass':
        tabela = tabela.drop(linha)

    #removendo os valores que não são lendários 
    elif lendario == False:
        tabela = tabela.drop(linha)

#mostrando a nova tabela 
display(tabela)
```

```python
# 2 . pegue o pokemon com mais vida 

#analisando linha por linha pegando a coluna HP
for linha in tabela_original.index:
    vida = tabela_original.loc[linha,'HP']

    #vendo o pokemon com mais vida
    max_vida = tabela_original['HP'].max()   
    
    #quando a linha analisada tiver com a maior vida
    if vida == max_vida: 

        #mostrando os resultados
        print(f"o pokemon com mais vida é {tabela_original.loc[linha,'Name']}")
```

```python
# 3 . pegue os pokemons com mais de 80 de spAtk e menos de 120 spAtk

#copiando a tabela original
tabela = tabela_original.copy()

#analisando linha por linha da tabela ccom a coluna Sp. Atk
for linha in tabela.index:
    spatk = tabela.loc[linha,'Sp. Atk']

    #removendo os valores que nao precisamos (<80 e >120)
    if spatk <= 80 or spatk >= 120:
        tabela = tabela.drop(linha)

#mostrando os resultados 
display(tabela)
```

```python
# 4 . pegue a geração pokemon com maior numero de pokemons

#numeros de pokemon por geração
tabela_valores = tabela_original['Generation'].value_counts().sort_index()

print(tabela_valores)

#observando qual geração pertencem e o maior numero de pokemons 
geracao = tabela_valores.idxmax()
num_pokemons = tabela_valores.max()

#mostrando resultados
print(f'\nO maior numero de pokemon esta na geração: {geracao} e são {num_pokemons} pokemons')
```

```python
# 5 . pegue os pokemons com apenas um único tipo, quantos pokemons são e qual a media entre os tipos

#copiando a tabela original e pegando os que não possuem 2º tipo
tabela = tabela_original[tabela_original['Type 2'].isnull()]

#contando os pokemons que possuem um único tipo
quant_nulo = tabela.shape[0]

#quantidade de tipos existentes 
quant_tipos = 18

#dividindo para saber a média
media_tipos = quant_nulo // quant_tipos

#mostrando resultados
display(tabela)
print(f'O numero de pokemons de tipo único são: {quant_nulo} e a media de tipos dentre eles é: {media_tipos}')
```

```python
# 6 .  pegue os pokemons que começam com j 

#copiando a tabela original
tabela_nomes= tabela_original.copy()

#utilizando a função que pega a primeira letra do nome dos pokemons
tabela = tabela_nomes[tabela_nomes['Name'].str.startswith('J')]

#mostrando os resultados com a letra desejada
display(tabela)
```

```python
# 7 . pegue o tipo mais comum de pokemon

#contando os tipos de pokemons 1 e 2 
contagem_tipos = tabela_original['Type 1'].value_counts()
contagem_tipos2 = tabela_original['Type 2'].value_counts()

#adicionando os dois resultados
contagem_geral = contagem_tipos.add(contagem_tipos2)

#pegando o maior numero 
tipo_mais_comum = contagem_geral.idxmax()

#mostrando os resultados 
print("O tipo de Pokémon mais comum é:", tipo_mais_comum)
```
